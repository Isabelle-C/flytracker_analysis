```python
rootdir = '/Volumes/Julie/2p-data'
session = '20240530'

basedir = os.path.join(rootdir, session, 'processed')
acqs = ['20240530_LC10a-SS1-GCaMP8m_f1v2_z1_TSeries-05302024-1610-001',
        '20240530_LC10a-SS1-GCaMP8m_f1v2_z1_TSeries-05302024-1610-002',
        '20240530_LC10a-SS1-GCaMP8m_f1v2_z1_TSeries-05302024-1610-003',
       '20240530_LC10a-SS1-GCaMP8m_f1v2_z1_TSeries-05302024-1610-004']
```

```python
t_list=[]
for acq in acqs:
    traces_fpath = os.path.join(basedir, '{}.csv'.format(acq))
    traces_ = pd.read_csv(traces_fpath)
    traces_ = traces_.rename(columns={'Mean': 'raw', 'Slice': 'frame'})
    traces_['acquisition'] = acq
    traces_['acqnum'] = acq.split('-')[-1]
    traces_['raw'] = traces_['raw'] - traces_['raw'].min()
    t_list.append(traces_)
    print(traces_.shape)
traces = pd.concat(t_list)
traces.head()
```

```python
fig, ax = pl.subplots(figsize=(10,5))
# ax.plot(traces[traces['acqnum']=='001']['raw'], 'r')
# ax.plot(traces[traces['acqnum']=='002']['raw'], 'orange')
ax.plot(traces[traces['acqnum']=='003']['raw'], 'b')
ax.plot(traces[traces['acqnum']=='004']['raw'], 'purple')
```
